```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# 1. Create SparkSession
spark = SparkSession.builder \
    .appName("test") \
    .config("spark.sql.shuffle.partitions", 2) \
    .getOrCreate()


df = spark.read.parquet('/output/*')
df.show(10)
```

```python
df.count()
```
